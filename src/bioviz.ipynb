```python
import pandas as pd
import rpy2.robjects as robj
from rpy2.robjects import numpy2ri
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr

deseq = importr("DESeq")
numpy2ri.activate()
pandas2ri.activate()

#datafile = R.system.file( "extdata/pasilla_gene_counts.tsv", package="pasilla" )
#d = {'a': robj.IntVector((11,12,13)), 'b': robj.IntVector((4,5,6))}
def normalize(countTable):
    conditions = robj.FactorVector( 'c' * len(countTable.columns) )
    cds = deseq.newCountDataSet( countTable, conditions )
    data = cds.slots['assayData']['counts']
    print(data)
    sizeFactors = deseq.estimateSizeFactorsForMatrix( data )
    return countTable / list(sizeFactors)
countTable = pd.DataFrame({'a': [11, 12, 13], 'b': [4, 5, 6]})
display(countTable)
display(normalize(countTable))
```
