```python
from sklearn.cluster import KMeans as KMeans_sklearn
import numpy as np

from vis.ge_scatter_tcr_graph import ScatterTCR

class KMeans:
    def __init__(self, batch, cols=['red', 'blue']):
        self.K = len(cols)
        self.clf = KMeans_sklearn(n_clusters=self.K, random_state=0).fit(batch.quant.ge)
        self.colors = pd.Series(data=map(lambda cell: cols[cell],
                                         self.clf.labels_.tolist()),
                                index=batch.index())
    
    def get_color(self, cell, annot=False):
        return self.colors[cell]
        
#batch.meta.set_color(colors)
#get_color = lambda cell, annot=False: batch.get_color(cell, coloring=Coloring.SAMPLE, annot=annot)
batch = run.B[('fluidigm', '')]
kmeans = KMeans(batch)
ScatterTCR(batch, dim_reduct='PCA', out_fn='/home/pesho/repos/bio-viz/out/fluidigm-special/fluidigm-2classes.png',
           get_color=kmeans.get_color)
```
